```python
#extract zip file

import zipfile

zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall()
zip_ref.close()
```

```python
#read data from file and handle null values

import pandas as pd

df = pd.read_csv('orders.csv', na_values = ['Not Available', 'unknown'])
# df.head(10)
```

```python
#rename column names from upper case to lower case

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
# df.head(5)
```

```python
#create new column for the discount amount given

df['discount'] = df['list_price']*df['discount_percent']*.01
# df.head(5)

df['sale_price'] = df['list_price']- df['discount']
df['profit'] = df['sale_price'] - df['cost_price']

# df.head(10)
```

```python
#convert order_data object to date

df['order_date'] = pd.to_datetime(df['order_date'],format = "%Y-%m-%d")
df.dtypes
df.head(2)
```

```python
# drop cost, list price and discount percent columns

# df.dtypes
df.drop(columns=['list_price', 'cost_price', 'discount_percent'],inplace=True)
```
